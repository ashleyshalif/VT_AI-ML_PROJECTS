```python
#nested dict
student = {
    101 : {"name" : "akash","age" : 26,"sem" : 5,"branch" : "cse" },
    102 : {"name" : "aka","age" : 26,"sem" : 5,"branch" : "cse" },
    103 : {"name" : "akashiiii","age" : 26,"sem" : 5,"branch" : "cse" }
}
for roll in student:
    #student[roll]["name"]
    s = student[roll]
    s["name"]
    s["age"]
```
