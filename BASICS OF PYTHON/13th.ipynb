```python
# function
''' - it is resuable block of code which is used to perform a specific task
- used to avoid repitition of code
- it is defined using def keyword
syntax : deffunction_name(parameters):
                statements'''
def name():
    print("welcome")
name()

# why do we use functions?
''' - increses readability of code
     -small code and easy to maintain and debug
     -boosts code resuability
'''
# How to pass parameters to function?
def greet(name,age):
    print(f"hello {age} {name}")

name1 = "ash"
greet(name1,34)
```

```python
#scope of variables
'''   variables
         |
    ------------
    |          |
    local    global
'''
global_variable = "this is global var"

def greet():
    print(global_variable)
    local_variable = "this is local var"
    print(local_variable)
greet()
```

```python
# function and keywords arguments
eggs = "human"
def spam():
    global eggs
    eggs = "chicken"
    print(eggs)
spam()
print(eggs)
```
