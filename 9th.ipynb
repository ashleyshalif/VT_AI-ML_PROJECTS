```python
'''set : built in structure
- stores unique value
-duplicate values not allowed
-indexing not suppported
-mutable(add,update)
-unordered collection
'''
nums = [1,3,4,3,1,5]
set(nums)
```

```python
A = {1,2,3}
B = {2,3,4}
C= {4,5,6}
print(A.union(B))
print(A.intersection(B))
#or
print(A&B)
print(A.union(B,C))
print(A.intersection(B,C))
```

```python
A = {1,2}
B = {1,2,3,4,5}
print((A^B))
print((A.difference(B)))
print((B.difference(A)))
#print((A.symmentic_difference(B)))
print(A.issubset(B))
print(A.issuperset(B))
```

```python
# 1)TAKE 10 INPUTS FROM USERS, OUTPUT SHOULD ONLY SHOULD UNIQUE NUMBERS
my_set = set()
for i in range (10) :
  num = int(input(f"enter num {i+1} :  "))
  my_set.add(num)
print(my_set)
```

```python
# 2)FROM THE INPUTED STRING, REMOVE DUPLICATE CHARACTERS
text = input("Enter a string: ")
unique_chars = "".join(sorted(set(text)))
print(f"String with unique characters: {unique_chars}")
```
