```python
''' QUESTION IS AS FOLLOWS :
====Employees management system ====
1. Parent classes:
              1.1  name
              1.2  salary
              1.3 display()

2. Child class: manager:
            2.1 department 
            2.2 display_manager()

Print details of employee and  manager '''
class InvalidSalaryError(Exception):
    pass
class Employee :
    def __init__(self) :
        self.name = input("PLEASE ENTER THE NAME OF EMPLOYEE : ")
        try :
            self.Salary = float(input("PLEASE ENTER THE SALARY OF THE EMPLOYEE : "))
            if self.Salary < 0:
                raise InvalidSalaryError("Salary cannot be negative.")
        except ValueError:
            print('SALARY HAS TO BE INPUTED IN FLOAT!! WRONG INPUT')

    def display(self):
        print(f"THE NAME OF THE EMPLOYEE IS : {self.name}")
        print(f"THE SALARY OF THE EMPLOYEE IS : {self.Salary}")
class Manager (Employee):
    def __init__(self):
        super().__init__()
        self.department = input('PLEASE ENTER THE DEPARTEMENT OF THE EMPLOYEE : ')
    def display_manager(self):
        print(f"THE DEPARTMENT OF THE MANAGER IS : {self.department}")
d = Manager()
d.display_manager()
d.display()

```
